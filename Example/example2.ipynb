```python
""" Set MChi to be 100 GeV """
MChi_GeV = 100
""" Set spectrum type """
type__ = 'p_mean'
""" Set cross section with nucleon"""
sigma0 = 1E-45 * CM2

""" Get SD rate function """
EventRate = XeSpinDependentRate(MChi_GeV, sigma0, type__)
```

```python
""" x axis """
X = np.linspace(0, 100, 100)
""" Corresponding recoil energy """
ER = X * KEV
""" Get event rate in events/day/keV/kg """
Y = [EventRate(Er) * DAY * KEV * KG for Er in ER]
```

```python
""" Visualization """
ax = plt.subplot(111)
ax.semilogy(X, Y)

ax.set_xlim([0, 100])
# ax.set_ylim([1E-9, 1E-4])
ax.set_xlabel('Recoil energy (keV)')
ax.set_ylabel('SD event rate (events/day/5keV/kg)')

""" Show the plot """
plt.show()
```
