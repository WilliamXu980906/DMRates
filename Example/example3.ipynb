```python
""" Set MChi to be 100 GeV """
MChi_GeV = 100
""" Set cross section with nucleon """
sigma0 = 1E-45 * CM2

""" Get event rate function """
EventRate = XeSpinIndependentRate(MChi_GeV, sigma0)
""" x axis """
X = np.linspace(0, 100, 100)
""" Corresponding recoil energy """
ER = X * KEV
""" Total event rate in events/ton/keV/year """
Y = [EventRate(Er) * YEAR * KEV * TON for Er in ER]
```

```python
""" Visualization """
ax = plt.subplot(111)
ax.semilogy(X, Y)

ax.set_xlim([0, 100])
ax.set_xlabel('Recoil energy (keV)')
ax.set_ylabel('SI event rate (events/year/keV/ton)')

""" Show the plot """
plt.show()
```
